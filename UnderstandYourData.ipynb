```python
# Create the first density plot for passengers who did not survive (Survived = 0)
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist=False, label='Not Survived')

# Create the second density plot for passengers who survived (Survived = 1)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist=False, label='Survived')

# Add a legend to differentiate the two density plots
plt.legend()

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
# Replace this with your own dataset or create sample data as needed
# For the purpose of the examples, we'll use some random data

# Sample Data for Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Sample Data for Scatter Plot
scatter_data = sns.load_dataset('iris')

# Sample Data for Bar Plot
bar_data = sns.load_dataset('tips')

# Sample Data for Histogram
hist_data = sns.load_dataset('iris')

# Sample Data for Box Plot and Violin Plot
box_violin_data = sns.load_dataset('tips')

# Sample Data for Pair Plot
pairplot_data = sns.load_dataset('iris')

# Sample Data for Heatmap
heatmap_data = sns.load_dataset('flights')

# 2.1 Line Plots
sns.lineplot(x=x, y=y)
plt.title('Line Plot')
plt.show()

# 2.2 Scatter Plots
sns.scatterplot(data=scatter_data, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot')
plt.show()

# 2.3 Bar Plots
sns.barplot(data=bar_data, x='day', y='total_bill', hue='sex')
plt.title('Bar Plot')
plt.show()

# 2.4 Histograms
sns.histplot(data=hist_data, x='sepal_length', kde=True)
plt.title('Histogram')
plt.show()

# 2.5 Box Plots
sns.boxplot(data=box_violin_data, x='day', y='total_bill', hue='sex')
plt.title('Box Plot')
plt.show()

# 2.6 Violin Plots
sns.violinplot(data=box_violin_data, x='day', y='total_bill', hue='sex', split=True)
plt.title('Violin Plot')
plt.show()

# 2.7 Pair Plots
sns.pairplot(pairplot_data, hue='species')
plt.suptitle('Pair Plot', y=1.02)
plt.show()

# 2.8 Heatmaps
heatmap_data = heatmap_data.pivot('month', 'year', 'passengers')
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='viridis')
plt.title('Heatmap')
plt.show()
```
